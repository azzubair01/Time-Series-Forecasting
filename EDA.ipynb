```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_ori = pd.read_csv('Dataset 2.csv', parse_dates=[0])
df_ori = df_ori.set_index('Date')
df_ori 
```

```python
df_ori.info()
```

```python
# Plot the available variables 
fig, axs = plt.subplots(4, 4)

axs[0, 0].plot(df_ori['FS'], 'tab:purple', marker ='o')
axs[0, 0].set_title('Food Sales')
axs[0, 0].grid()
axs[0, 0].set_ylabel('RM (million)')

axs[0, 1].plot(df_ori['ICG'], 'tab:red', marker ='o')
axs[0, 1].set_title('Import of Consumption Goods')
axs[0, 1].grid()
axs[0, 1].set_ylabel('RM (million)')

axs[0, 2].plot(df_ori['TSP'], 'tab:blue', marker ='o')
axs[0, 2].set_title('Television Set Production')
axs[0, 2].grid()
axs[0, 2].set_ylabel('Units')

axs[0, 3].plot(df_ori['SOM'], 'tab:orange', marker ='o')
axs[0, 3].set_title('Sales of Motorcycle')
axs[0, 3].grid()
axs[0, 3].set_ylabel("Units ('000)")

axs[1, 0].plot(df_ori['SOPC'], 'tab:green', marker ='o')
axs[1, 0].set_title('Sales of Passenger Cars')
axs[1, 0].grid()
axs[1, 0].set_ylabel("Units ('000)")

axs[1, 1].plot(df_ori['CC'], 'tab:pink', marker ='o')
axs[1, 1].set_title('Loans disbursed for Credit Consumption')
axs[1, 1].grid()
axs[1, 1].set_ylabel('RM (million)')

axs[1, 2].plot(df_ori['CCD'], 'tab:red', marker ='o')
axs[1, 2].set_title('Consumption Credit Disbursed')
axs[1, 2].grid()
axs[1, 2].set_ylabel('RM (million)')

axs[1, 3].plot(df_ori['CCS'], 'tab:cyan', marker ='o')
axs[1, 3].set_title('Credit Consumption Spending')
axs[1, 3].grid()
axs[1, 3].set_ylabel('RM (million)')
                     
axs[2, 0].plot(df_ori['LDH'], 'tab:grey', marker ='o')
axs[2, 0].set_title('Loans disbursed to Household Sector')
axs[2, 0].grid()
axs[2, 0].set_ylabel('RM (million)')

axs[2, 1].plot(df_ori['LOWT'], 'tab:purple', marker ='o')
axs[2, 1].set_title('Loans disbursed on Whole Trade, Retail, & Hotel')
axs[2, 1].grid()
axs[2, 1].set_ylabel('RM (million)')

axs[2, 2].plot(df_ori['MSW'], 'tab:red', marker ='o')
axs[2, 2].set_title('Manufacturing Salaries and Wages')
axs[2, 2].grid()
axs[2, 2].set_ylabel("RM (million)")

axs[2, 3].plot(df_ori['MIER'], 'tab:blue', marker ='o')
axs[2, 3].set_title('MIER Consumer Sentiment Index')
axs[2, 3].grid()
axs[2, 3].set_ylabel('Consumer Sentiment Index')

axs[3, 0].plot(df_ori['MCF'], 'tab:orange', marker ='o')
axs[3, 0].set_title('Market Capitalisation of FBM KLCI')
axs[3, 0].grid()
axs[3, 0].set_ylabel('RM (million)')

axs[3, 1].plot(df_ori['EMP'], 'tab:green', marker ='o')
axs[3, 1].set_title('Number of Employed Workers')
axs[3, 1].grid()
axs[3, 1].set_ylabel('Unit')

axs[3, 2].plot(df_ori['NM'], 'tab:pink', marker ='o')
axs[3, 2].set_title('Narrow Money (M1)')
axs[3, 2].grid()
axs[3, 2].set_ylabel('RM (million)')

axs[3, 3].plot(df_ori['FBM'], 'tab:red', marker ='o')
axs[3, 3].set_title('FTSE Bursa Malaysia KL Composite Index')
axs[3, 3].grid()
axs[3, 3].set_ylabel('KL Composite Index')


fig.subplots_adjust(left=1, bottom=3, right=5, top=7, wspace=None, hspace=None)
```
